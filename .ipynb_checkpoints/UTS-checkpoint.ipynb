```python
# Data
company_detail_list = [
    {'name': 'Company 1', 'domain': 'Retail', 'country': 'United States'},
    {'name': 'Company 2', 'domain': 'Technology', 'country': 'United Kingdom'},
    {'name': 'Company 3', 'domain': 'Healthcare', 'country': 'United States'}
]

employee_detail_list = [
    {'name': 'EMP-0001', 'first_name': 'John', 'last_name': 'Doe', 'full_name': 'John Doe', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0002', 'first_name': 'Tom', 'last_name': 'Smith', 'full_name': 'Tom Smith', 'company': 'Company 2', 'nationality': 'United States'},
    {'name': 'EMP-0003', 'first_name': 'Andrew', 'last_name': 'Sebastian', 'full_name': 'Andrew Sebastian', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0005', 'first_name': 'Ying Han', 'last_name': 'Tan', 'full_name': 'Ying Han Tan', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0015', 'first_name': 'Kenneth', 'last_name': 'Ng', 'full_name': 'Kenneth Ng', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0018', 'first_name': 'Rubby', 'last_name': 'Lee', 'full_name': 'Rubby Lee', 'company': 'Company 2', 'nationality': 'Hong Kong'},
    {'name': 'EMP-0017', 'first_name': 'Robert', 'last_name': 'White', 'full_name': 'Robert White', 'company': 'Company 1', 'nationality': 'United Kingdom'}
]

# Task 1: Mendapatkan daftar semua perusahaan dan mengurutkannya secara terbalik berdasarkan nama perusahaan
def get_sorted_companies():
    sorted_companies = sorted(company_detail_list, key=lambda x: x['name'], reverse=True)
    return [{'name': company['name']} for company in sorted_companies]

# Task 2: Menampilkan setiap nilai "Domain" untuk setiap perusahaan
def print_company_domains():
    for company in company_detail_list:
        print(f"{company['name']}: {company['domain']} ({company['country']})")

# Task 3: Mendaftar semua karyawan berdasarkan domain perusahaan
def list_employees_by_domain():
    domain_employees = {}
    for company in company_detail_list:
        domain_employees[company['domain']] = []
    for employee in employee_detail_list:
        company = next(c for c in company_detail_list if c['name'] == employee['company'])
        domain_employees[company['domain']].append(employee['full_name'])
    return domain_employees

# Task 4: Membuat fungsi yang mengembalikan karyawan beserta negara perusahaan mereka
def get_employees_with_company_country():
    employees_with_country = []
    for employee in employee_detail_list:
        company = next(c for c in company_detail_list if c['name'] == employee['company'])
        employees_with_country.append({
            "full_name": employee['full_name'],
            "company": employee['company'],
            "country": company['country']
        })
    return employees_with_country

# Task 5: Membuat fungsi yang menghitung jumlah karyawan berdasarkan kewarganegaraan di tiap perusahaan
def count_employee_nationalities():
    company_nationality_counts = []
    for company in company_detail_list:
        nationality_count = {}
        employees = [e for e in employee_detail_list if e['company'] == company['name']]
        for employee in employees:
            nationality = employee['nationality']
            nationality_count[nationality] = nationality_count.get(nationality, 0) + 1
        company_nationality_counts.append({
            "company": company['name'],
            "employee_nationality": nationality_count
        })
    return company_nationality_counts


# Task 1
print("Task 1 Result:")
for company in get_sorted_companies():
    print(company)

# Task 2
print("\nTask 2 Result:")
print_company_domains()

# Task 3
print("\nTask 3 Result:")
domain_employees = list_employees_by_domain()
for domain, employees in domain_employees.items():
    print(f"{domain}:")
    for employee in employees:
        print(f"    - {employee}")

# Task 4
print("\nTask 4 Result:")
employees_with_country = get_employees_with_company_country()
for employee in employees_with_country:
    print(employee)

# Task 5
print("\nTask 5 Result:")
company_nationality_counts = count_employee_nationalities()
for company in company_nationality_counts:
    print(f"{company['company']}:")
    for nationality, count in company['employee_nationality'].items():
        print(f"    {nationality}: {count}")
```
